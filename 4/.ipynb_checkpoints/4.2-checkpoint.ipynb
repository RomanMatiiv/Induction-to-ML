```python
import pandas as pd
from datetime import datetime
```

```python
from sklearn.decomposition import PCA
from numpy import corrcoef
```

```python
data=pd.read_csv("close_prices.csv")
index=pd.read_csv("djia_index.csv")
```

```python
data.date=[datetime.strptime(x,"%Y-%m-%d") for x in data.date.values]
```

```python
data=data.iloc[0:,1:]
```

```python
data.head()
```

```python
pca=PCA(n_components=10)
```

```python
pca.fit(X=data)
```

```python
sum=0
for x in pca.explained_variance_ratio_:
    print x
    sum+=x
    print sum
    print
```

```python
with open("a1.txt","w") as f:
    f.write("4")
```
